# Estruturas de Repetição (for, while)

## Comando for

* for (variável) in (sequência):
*          (instrução identada)
* (bloco de código não identdo)

## Formas de Repetição

* Range
* Simples
* Enumerate



```python
# for Range
for contador in range(10): # de 0 até i-1
  print(contador)
```

```python
# for Range + Break
for contador in range(10):
  print(contador)
  if contador == 2:
    break
```

```python
lista = ['Teste1', 'Teste2']
```

```python
# for Simples
for var in lista:
  print(var)
```

```python
# for com Enumerate
for i, val in enumerate(lista):
  print(i, var)
```

## for com Opedadores Compostos

```python
lista2 = [1, 2, 3, 4]
```

```python
for var in lista2:
  var += 1 # var = var + 1
  print(var)
```

```python
for var in lista:
  var *= 2 # var = var * 2
  print (var)
```

## Comando While

* while (condição):
*       (instrução identada)
* (bloco de código não identado)

```python
var = 1
while var < 6:
  var *= 2
  print(var)
```

```python
# while + else
var = 1
while var < 6:
  var *= 2
else:
  print('O valor de var é', var)
```
