# Estruturas Condicionis
* if
* if/else
* if/elif/else

* Condicional

```python
# if
num1 = eval(input('Digite o primeiro número '))
num2 = eval(input('Digite o segundo número '))

if num1 > num2:
  print('O num1 é o maior número')
```

```python
# if/else
num1 = eval(input('Digite o primeiro número '))
num2 = eval(input('Digite o segundo número '))

if num1 > num2:
  print('O num1 é o maior númeri')
else:
  print('O num2 é o maior número')
```

```python
# if/elif/else
num1 = eval(input('Digite o primeiro número '))
num2 = eval(input('Digite o segundo número '))

if num1 > num2:
  print('O num1 é o maior númeri')
elif num1 < num2:
  print('O num2 é o maior número')
else: # em uma estrutura condicional de três vias ou mais o comando else é opcional (poderia ser mais um elif)
  print('Os dois são iguais')
```

* Condicional com Operdores Lógicos

```python
num1 = eval(input('Digite o primeiro número '))
num2 = eval(input('Digite o segundo número '))
num3 = eval(input('Digite o terceiro número '))

if (num1 > num2) and (num1 > num3):
  print('O num1 é o maior número')
else:
  print('O num1 não é o maior número')
```
