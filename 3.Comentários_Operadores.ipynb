# Operadores Aritméticos
## Resultam em um tipo int ou um tipo float

```python
# Adição
10 + 2
```

```python
# Subtração
10 - 2
```

```python
# Multiplicação
10 * 2
```

```python
# Divisão
10 / 2
```

```python
# Divisão Inteira
10 // 2
```

```python
# Potenciação
10 ** 2
```

# Operadores Relacionais
## Resultam em um tipo bool - True or False

```python
# Menor
10 < 2
```

```python
# Maior
10 > 2
```

```python
# Menor ou Igual
10 <= 2
```

```python
# Maior ou Igual
10 >= 2
```

```python
# Igual
10 == 2
```

```python
# Diferente
10 != 2
```

# Operadores Lógicos
## And
## Or
## Not

## Tabela Verdade - And

```python
print(10 > 2 and 4 < 6) # True and True -> True
print(10 > 2 and 4 > 6) # True and False -> False
print(10 < 2 and 4 < 6) # False and True -> False
print(10 < 2 and 4 > 6) # False and Flase -> False
```

# Tabela Verdade - Or

```python
print(10 > 2 or 4 < 6) # True or True -> True
print(10 > 2 or 4 > 6) # True or False -> False
print(10 < 2 or 4 < 6) # False or True -> True
print(10 < 2 or 4 > 6) # False or Flase -> False
```

## Tabela Verdade - Not

```python
print(not(10 > 2)) # True -> False
print(not(10 < 2)) # False -> True
```
