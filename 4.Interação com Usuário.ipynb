# Type

```python
x = 10 # x recebe o inteiro 10
type(x) # função type - retorna qual é o tipo do dado
print('x', type(x))

nome = 'Rodrigo'
print('nome', type(nome))

num = '100' # num recebe a STRING 100
num_convert = int(num) # função construtora int - converte o dado em um tipo int
print('num', type(num))
print('num_convert', type(num_convert))

num_float = float(num) # função construtora float - converte o dado em um tipo float
print('num_float', num_float)

```

# Input
* Armazenar a informação digitada pelo usuáruo em uma variável
* O dado armazenado sempre será um texto


```python
nume = input('Digite o nome do usuário ') # Aguarda o usuário digitar algo
print(nome)
```

```python
valor = input('Digite a nota ')

print(nome, 'sua nota foi', valor)
```

```python
valor = input('Digite o valor do produto ')
print(valor)
```

```python
desconto = valor * 0.15
print(desconto)
```

```python
# Primeira solução - utilizar a função construtora int (se o valor digitado for um número inteiro)
valor_convert = int(valor)
desconto = valor_convert * 0.15
print(desconto)
```

```python
# Segunda solução - utilizar a funççao eval, envolvendo a função print
valor = eval(input('Digite o valor do produto ')) # A função eval transforma todo dado recebido, em input, em um valor numérico
desconto = valor * 0.15
print(valor)
print(desconto)
```
