# Regras de Variáveis

**Nome da Variável**
* Começar com letra (a-z, A-Z) ou underscore(_)
* São Case Sensitive (Diferença entre letras maiúsculas ou minúsculas)
> Ex: A != a
* Usar nomes que definam para que a variável serve no programa
* Não usar nomes extensos
> Ex: Soma de dois números
* a = 2
* b = 5
* soma = 2 + 5 (Boa prática)
* soma_total = 2 + 5 (Boa Prática)

# Tipos de Variáveis

**Python é uma linguagem de alto nível**
* Não é necessário especificar o tipo de dados quando declarar a variável
* Python define o tipo do dado, automaticamente, na declaração
> EX: Tipos de Dados
* Int (inteiro) : num = 10
* Float (Decimal): num = 10.5 - o separador decimal é o ponto (.)
* String (Texto): nome = 'Rodrigo' - aspas duplas ou simples
* Bool (booleano): True or False
