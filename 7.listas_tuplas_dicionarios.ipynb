# Listas
* Uma lista é uma sequência de objetos
* Os objetos podem ser de diferentes tipos: números (int, float), strings (str) e até mesmo outras listas

```python
lista = [5, 'Rodrigo', 9.8]
```

```python
type(lista)
```

```python
list.append('Augusto') # Adicionar um elemento ao final da lista
```

```python
print(list)
print(list[1])
```

# Tuplas
* Tuplas são como as listas, porém imutáveis


```python
tupla = (5, 'Rodrigo', 9.8)
```

```python
type(tupla)
```

```python
tupla.append('Augusto')
```

```python
print(tupla)
```

# Dicionário
* chave: valor

```python
dicionario = {'nome': 'Rodrigo', 'cidade': 'São José Dos Campos'}
```

```python
type(dicionario)
```

```python
print(dicionario)
```

```python
print(dicionario['nome'])
```

```python
dicionario['ultimo_nome'] = 'Rebouças' # não é permitido usar o método append em dicionários Python
```

```python
print(dicionario)
```
